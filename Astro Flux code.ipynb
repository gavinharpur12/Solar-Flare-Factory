```python
import matplotlib.pyplot as plt
import numpy as np

from astropy.visualization import time_support

from sunpy import timeseries as ts
from sunpy.net import Fido
from sunpy.net import attrs as a
```

```python
# Start and stop time to search for files from the XRS instrument aboard the GOES satellite
tstart = "2022-03-30 11:59"
tend = "2022-03-30 23:59"
result = Fido.search(a.Time(tstart, tend), a.Instrument("XRS"))
print(result)
```

```python
# Print only the files from the GOES 17 satellite

result_goes17 = Fido.search(a.Time(tstart, tend), a.Instrument("XRS"), a.goes.SatelliteNumber(17), a.Resolution("flx1s"))
print(result_goes17)
```

```python
file_goes17 = Fido.fetch(result_goes17)
```
